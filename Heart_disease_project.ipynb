```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from ipywidgets import widgets

%matplotlib inline
```

```python
X = pd.read_csv(r'path', index_col='patient_id')
```

```python
#plot possible dependencies between the features
sm = pd.plotting.scatter_matrix(X, alpha=0.2, figsize=(20, 20))

#Change label rotation
[s.xaxis.label.set_rotation(90) for s in sm.reshape(-1)]
[s.yaxis.label.set_rotation(0) for s in sm.reshape(-1)]

[s.get_yaxis().set_label_coords(-1,0.5) for s in sm.reshape(-1)]

#Hide all ticks
[s.set_xticks(()) for s in sm.reshape(-1)]
[s.set_yticks(()) for s in sm.reshape(-1)]
plt.savefig('fig2.pdf', bbox_inches='tight')
```

```python
#the figure is the same as above but shows the heat map based on linear correlation

plt.matshow(X.corr())
plt.xticks(range(X.shape[1]), X.columns, rotation=90)
plt.yticks(range(X.shape[1]), X.columns)
plt.colorbar()

plt.savefig('fig3.pdf', bbox_inches='tight')
plt.show();
```
